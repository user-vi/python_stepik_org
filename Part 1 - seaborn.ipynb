```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/iris.csv')
df.head()
```

```python
sns.distplot(df['sepal length'], kde_kws={"label": "sepal length"})
sns.distplot(df['sepal width'], kde_kws={"label": "sepal width"})
sns.distplot(df['petal length'], kde_kws={"label": "petal length"})
sns.distplot(df['petal width'], kde_kws={"label": "petal width"})
```

```python
sns.kdeplot(df['sepal length'])
sns.kdeplot(df['sepal width'])
sns.kdeplot(df['petal length'])
sns.kdeplot(df['petal width'])
```
