```python
df = pd.read_csv("https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv")
df.head(3)
```

```python
#фильтрация строк
df.query("gender == 'female' & lunch == 'standard'").head()
```

```python
#фильтрация строк. query не работает с пробелами, слешами в названиях колонок
df = df.rename(columns = {'math score': 'math_score'})
test_math_score = 40
df.query("gender == 'female' & math_score < @test_math_score").head(3)
```

```python
#фильтрация. axis = 1 для колонок, axis = 0 для строк
df.filter(like = "score", axis = 1).head()
```

```python
#фильтрация. сделать subset
df.filter(like = "score", axis = 1).query("math_score < 60").head(5)
```

```python
#группировка и аггрегация

#df.groupby('gender').mean() или
df.groupby(['gender', 'race/ethnicity'], as_index = False)\
    .aggregate({'math score':'mean', 'reading score':'sum'})\
    .rename(columns = {'math score':'mean math score', 'reading score':'sum reading score'})
```

```python
#сортировка
df.sort_values(['gender', 'math score'], ascending = False) \
    .groupby('gender').head(3)
```

```python
#добавление столбца
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df.head(5)
```

```python
#удаление столбца
df.drop(['total_score'], axis = 1).head()
```
